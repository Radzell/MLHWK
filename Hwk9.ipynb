```python
import pandas as pd
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
import math
from scipy import linalg
from sklearn.cross_validation import ShuffleSplit
from sklearn import linear_model
```

```python
def calc_err(Z, y, w):
        y_hat = np.sign(np.dot(Z, w))
        y_diff = [4 if i[0] != i[1] else 0 for i in zip(y, y_hat)]
        err = sum(y_diff)/float(len(y))
        return err
def legendre(x, n):
        """ Returns the nth legendre polynomial of x """
        L = []
        for i in range(n+1):
                if i == 0: 
                        L.append(1)
                        continue
                if i == 1: 
                        L.append(x)
                        continue

                k = float(i)
                Li = (2*k-1)*x*L[i-1] / (k) - (k-1)*L[i-2] / (k)
                L.append(Li)
        
        
        return L[n]

                        
def polynomial_transform(x, n):
        """ Returns the nth order legendre polynomial transform of x """
        x1 = x2 = 0
        z = []
        if len(x) == 3:
                x1 = x[1]
                x2 = x[2]
        if len(x) == 2:
                x1 = x[0]
                x2 = x[1]
        
        for i in range(n+1):
                for j in range(n+1):
                        if (i+j) > n: continue
                        z1 = legendre(x1, i)
                        z2 = legendre(x2, j)
                        z.append(z1*z2)
        
        return z
def symmetryfunc(row):

    img = row[1:]
    sym = float(0)
    for i in range(len(img)/2):
        row = i/16
        col = 16 - (i % 16)
        
        com = 256 - (16 * row) - col
        
        
        sym += abs(img[i] - img[com])
    return sym
def displaydata(df):
    plt.scatter(df[(df[0] == 1)]['ints'],df[(df[0] == 1)]['sym'],marker = 'o',c = 'b',label='5s')
    plt.scatter(df[(df[0] == 5)]['ints'],df[(df[0] == 5)]['sym'],marker = 'x',c = 'r',label='1s')
    plt.legend()
    plt.xlabel('Intensity')
    plt.ylabel('Symmetry')
    plt.ylim([-1,1])
    plt.xlim([-1,1])
    plt.show()
def createweights2(a,b,lamb):
    a = np.array(a.tolist())
    b = np.array(b.tolist())
    return linalg.inv(a.T.dot(a)+lamb*np.identity(len(a.T))).dot(a.T).dot(b)
def createweights(df,lamb):
    a = np.array(df.ix[traincv]['transform'].tolist())
    b = np.array(df.ix[traincv]['target'].tolist())
    return linalg.inv(a.T.dot(a)+lamb*np.identity(len(a.T))).dot(a.T).dot(b)
def plot_bound(combinedset,mask,lamb):
    t=createweights(combinedset.ix[traincv],lamb)
    #plot the weights
    density = 100
    xlist = np.linspace(-1.0,1.0,density)
    ylist = np.linspace(-1.0,1.0,density)
    pospoints = []
    negpoints = []
    all_points = [(x,y) for x in xlist for y in ylist]
    result = []
    all_grid = [all_points[i:i+density] for i in range(0, len(all_points), density)]
    for i in range(density):
        for j in range(density):
            z = polynomial_transform(all_grid[i][j], 8)
            classof = np.sign(np.dot(t, z))
            if classof == 0:
                print hey
            if classof == 1:
                    pospoints.append(all_grid[i][j])
            else:
                    negpoints.append(all_grid[i][j])
    plt.scatter(zip(*pospoints)[0], zip(*pospoints)[1], marker='o', color='m')
    plt.scatter(zip(*negpoints)[0], zip(*negpoints)[1], marker='o', color='g')
    displaydata(combinedset.ix[mask])
def cross_val(Z, y, lamb):
        """ Returns the cross validation error for linear reg with weight decay """
        
        N = len(y)
        wreg = createweights2(Z, y, lamb)
        y_hat = np.sign(np.dot(Z, wreg))
        ZT = np.transpose(Z)
        ZTZ = np.dot(ZT, Z)
        L = lamb * np.identity(len(ZTZ))
        ZTZL = ZTZ + L
        inv = np.linalg.inv(ZTZL)
        pseudoinv = np.dot(inv, ZT)
        pseudoinvT = np.transpose(pseudoinv)

        H_ish = np.multiply(Z, pseudoinvT)
        H_diag = [sum(i) for i in H_ish]

        

        indiv_err = sum(((y_hat[i] - y[i])/(1-H_diag[i]))**2 for i in range(N))
        Ecv = indiv_err
        Ecv /= N

        return Ecv
```

```python
trainset = pd.read_csv(os.getcwd()+"/ZipDigits.train",header=None,delim_whitespace=True)
testset = pd.read_csv(os.getcwd()+"/ZipDigits.test",header=None,delim_whitespace=True)
#remove all numbers but the 1s and the 5s
trainset = trainset[(trainset[0] == 1) | (trainset[0] == 5)]
testset = testset[(testset[0] == 1) | (testset[0] == 5)]

for col in range(1,len(trainset.columns)):
    trainset[col]+=1.0
    trainset[col]*=128
for col in range(1,len(testset.columns)):
    testset[col]+=1
    testset[col]*=128
    
#do sym and int for the sets
trainset['sym'] =None
trainset['sym'] = trainset.apply(lambda row: symmetryfunc(row), axis=1)
trainset['ints'] =  trainset.apply(lambda row: sum(row[1:])/len(row[1:]), axis=1)

#do sym and int for the sets
testset['sym'] =None
testset['sym'] = testset.apply(lambda row: symmetryfunc(row), axis=1)
testset['ints'] =  testset.apply(lambda row: sum(row[1:])/len(row[1:]), axis=1)

#combine the dataset
combinedset = pd.concat([trainset,testset],ignore_index=True)

#normilize the combinedset
combinedset['ints'] -=combinedset['ints'].max()/2
combinedset['sym'] -=combinedset['sym'].max()/2
combinedset['sym'] /=max( combinedset['sym'].max(),math.fabs(combinedset['sym'].max()))
combinedset['ints'] /=max( combinedset['ints'].max(),math.fabs(combinedset['ints'].max()))
#transform the combinedset
combinedset['transform']=combinedset.apply(lambda row: polynomial_transform([row['ints'],row['sym']],8), axis=1)
combinedset['target']=combinedset[0].map(lambda x: 1 if x == 5 else -1)
```
