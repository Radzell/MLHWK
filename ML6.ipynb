```python
import random
from random import choice
import numpy as np
from __future__ import division
from matplotlib import pyplot
import math

sep = 2
thk = 1

topOriginY = sep/2.0
topOriginX = -2

botOriginY = -(sep/2.0)
botOriginX = 2

rad1 = 3
rad2 = 4

topy = []
topx = []

boty = []
botx = []

points = []
points2 = []
yans = []

for i in range(500):
    randrad = random.uniform(rad1,rad2)
    theta = random.uniform(0,(math.pi))
    
    x = (randrad * math.cos(theta)) + topOriginX
    y = (randrad * math.sin(theta)) + topOriginY
    
    topx.append(x)
    topy.append(y)
    
    points.append([1,x,y,1])
    points2.append([1,x,y])
    yans.append(1)
    
    
for i in range(500):
    randrad = random.uniform(rad1,rad2)
    theta = random.uniform(math.pi,(2 * math.pi))
    
    x = (randrad * math.cos(theta)) + botOriginX
    y = (randrad * math.sin(theta)) + botOriginY
    
    botx.append(x)
    boty.append(y)
    
    points.append([1,x,y,-1])
    points2.append([1,x,y])
    yans.append(-1)
    

weights = np.array([.1,.1,.1])

def isLeft(weights, point):
    return np.dot(np.transpose(weights),point[:3]) > 0

count = 0        
j = 0
while j < len(points): # updating perceptron g 
    
    i = points[j]
    
    if ((isLeft(weights,i)) and (i[3] == -1)) or ((not isLeft(weights,i)) and (i[3] == 1)):
        
        
        x = np.array(i[:3])
        y = i[3]
        
        weights += (y*x)
        
        j=0
        
        count +=1
    else:
        j+=1
        
interceptg = np.array([-weights[0],weights[2]])       # seeing if we succeded

slopeg = np.array([-weights[1],weights[2]])
```
